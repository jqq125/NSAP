(包括：ut_edges.npy,us_edges.npy,sg_edges.npy)

```python
import csv
import numpy as np
import xlwt #进行excel数据
```

```python
file='../gs_edges_id2newid.txt'
file.split("/")
```

```python
file.split("/")[1].split('_')[0]
```

```python
#获取drug-target原始边的编号
edges=[]
type= file.split("/")[1].split('_')[0]
filename=type+'_edges.npy'
print(filename)
```

```python
ut_edges=[]
with open(file,'r') as f:
    reader=f.readlines()
    n=len(reader)
    for i in range(n):
        row=reader[i]
        #print(row)
        data=str(row).strip('\n').split("\t")
        #print(data)
        drug=int(data[1])
        target=int(data[3])
        ut_edges.append([drug,target])
    print(ut_edges)
        #break
    print(len(ut_edges))
    np.save(filename,ut_edges)
```
